```python
import pandas as pd #pandas untuk manipulasi data
import seaborn as sns #untuk visualisasi data
import matplotlib.pyplot as plt #untuk visualisasi data

df_produk = pd.read_csv('produkpalinglaku.csv')

# Cek nama kolom (opsional)
print(df_produk.columns)

plt.figure(figsize=(8, 5))
sns.barplot(data=df_produk, x='total_sales', y='productline', palette='Blues_d')
plt.title("Top 5 Product Line dengan Penjualan Tertinggi")
plt.xlabel("Total Penjualan")
plt.ylabel("Product Line")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Baca CSV kamu
df = pd.read_csv('penjualanperbulan.csv')

# 2. Ubah kolom bulan ke format datetime
df['bulan'] = pd.to_datetime(df['bulan'])

# 3. Bikin label bulan yang rapi (Jan 2003, Feb 2003, dst)
df['bulan_label'] = df['bulan'].dt.strftime('%b %Y')

# 4. Visualisasi Line Chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='bulan', y='total_penjualan', marker='o')
plt.xticks(ticks=df['bulan'], labels=df['bulan_label'], rotation=45)
plt.title("Tren Total Penjualan per Bulan")
plt.xlabel("Bulan")
plt.ylabel("Total Penjualan")
plt.grid(True)
plt.tight_layout()
plt.show()
```
